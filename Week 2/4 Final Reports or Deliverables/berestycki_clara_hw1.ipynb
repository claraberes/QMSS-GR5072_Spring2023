This project uses _deep convolutional neural networks_ to predict __deforestation patterns in the amazon__. The authors combine 3 different datasets as base data. The developped algorithm takes into the time dimension: with information on deforestation patterns for 10 years, the model accurately predicts deforestatioon patterns at time t+1. 

The github repository for this project is [here](https://github.com/PatBall1/DeepForestcast) and the link to the associated article [here](https://doi.org/10.1101/2021.12.14.472442)

Machine learning is useful in this set up because it is hard to get ground truth on deforested parcels. Deforestation occurs in very remote areas. 

## Data sources 

* The first dataset is the Global Forest Change dataset published by Hansen et. al in 2013. This dataset records forest loss at the 30x30m resolution from 2001 to 2020 (the dataset is updated annually). This dataset provides the labels. 

* The second source of information for the model is Landsat, or satellite imagery, from year 2014 to 2020.

* The last dataset is Japan Aerospace Exploration Agency's (JAXA) 30 m resolution ALOS Global DSM which gives topological information abotu the area. 

## Models

The authors first try a 2D CNN, but with this model they can only include the (1) spatial and (2) spectral dimensions. They try a 3D CNN to add the __time__ dimension, which can be expected to be crucial because of the nature of the project. 

Model hyperparameters are set on the 2D CNN to gain time. For the optimiser, they use the standard Adams optimisier. For the loss function, they use BCEWITHLOGITSLOSS, which is a loss function specific to binary classification problem (this is the case here as the outcome is either "forested" or "deforested". The image below represents the evolution of the training and validation losses (in machine learning, datasets are typically split into a training set (where the model "learns"), a validation set to finetune, and a test set. 

There are very few deforested pixels compared to forested pixels in the overall dataset. The specificity of this problem is that in order for the model to perform well, more weight has to be given to __actual__ deforested pixels in the training set. Otherwise the model won't see enough to perform well in the test set. 





The last category of models are recurrent neural networks. This type of model is specifically fitted for sequential data. 

## Results 

The Convolutional RNN on github has the least accuracy (Accuracy meassure=0.66) while the 2D and 3D CNN are better with accuracy results above 0.97. The accuracy measure here is the AUC or area under the curve. Here it means that 97% of the time, the model is able to predict whether a given pixel will be deforested at year t+1.
